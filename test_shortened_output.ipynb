```python
x = pickle5.load(open('test_data/x_mini.pkl', 'rb'))
x_values = list(x.values())
y = pickle5.load(open('test_data/y_mini.pkl', 'rb'))
y_values = list(y.values())
```

```python
y_rows = []
x_rows = []
for i in range(len(x_values)):

    idx = statistics.get_masked(x_values[i],10)

    y_rows.append(y_values[i][idx].reshape(1,10))
    x_rows.append(x_values[i][:,:10])

x_rows_dict = dict(zip(list(x.keys()), x_rows))
y_rows_dict = dict(zip(list(y.keys()), y_rows))

pickle5.dump(y_rows_dict, open('test_data/y_mini_row.pkl', 'wb'))
pickle5.dump(x_rows_dict, open('test_data/x_mini_row.pkl', 'wb'))
```
