```python
base = '/home/grig0076/phispy_phrog_pickles/cross_validated/models/all_features/'
ROC_mat = [[] for i in range(10)]
#colors = ["#332288", "#88CCEE", "#44AA99", "#117733", "#999933", "#DDCC77", "#CC6677", "#882255", "#AA4499"] 
colors = ["#1b9e77", "#d95f02", "#7570b3", "#e7298a", "#66a61e","#cab2d6",  "#a6761d", "#e6ab02", "#80b1d3"]
for i in range(10): 
    
    file_name = base + 'prophage_phrog_data_derep_fourormore_lessthan121_' + str(i) + '_all_chunk_trained_LSTMbest_val_loss_ROC.tsv'
    df = pd.read_csv(file_name, sep = '\t', index_col = 'Unnamed: 0')
    
    categories = [i for i in df.columns.to_list() if i != 'FPR']
    
    for j in range(len(categories)): 
        
        ROC_mat[j].append(df[categories[j]].to_list())  
    
FPR = df['FPR']
sns.set(font_scale=2)

sns.set_style("dark")
plt.rcParams["figure.figsize"] = (10,10)

for k in range(len(categories)): 
    
    this_df = pd.DataFrame(np.array(ROC_mat[k]).T)
    mean = this_df.mean(axis = 1)
    #std = 1.96 * this_df.std(axis = 1)/np.sqrt(4)
    std =  this_df.std(axis = 1)
    upper = mean + std
    lower = mean - std
    
    ax = plt.plot(FPR, mean,  color = colors[k], label = categories[k], lw = 3.5)
    plt.fill_between(FPR,lower, upper, color = colors[k], alpha = 0.2) 
    
plt.plot([0, 1], [0, 1], color="black", lw=2, linestyle="--")

    
plt.xlabel('False Positive Rate')
plt.ylabel('False Negative Rate')
plt.savefig('/home/grig0076/GitHubs/Phynteny/all_features_ROC.png')    

leg = plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
# change the line width for the legend
for line in leg.get_lines():
    line.set_linewidth(12.0)
    
plt.show() 
```

### Metrics

```python
def error_margin(values): 
    """ 
    Get the margin of error, to be used for calculating confidence intervals 
    
    :param values: List of values to calculate margin of error for 
    :return: margin of error  
    """
    
    return np.std(values)/np.sqrt(len(values)) 
```

```python
base = '/home/grig0076/phispy_phrog_pickles/cross_validated/models/all_features/'

precision_list = [] 
recall_list = [] 
f1_list = [] 

for i in range(10): 
    
    file_name = base + 'prophage_phrog_data_derep_fourormore_lessthan121_' + str(i) + '_all_chunk_trained_LSTMbest_val_loss_metrics.pkl'
    
    this_metrics = pickle5.load(open(file_name, 'rb'))
    
    categories = this_metrics['category']
    precision_list.append(this_metrics['precision'])
    recall_list.append(this_metrics['recall'])
    f1_list.append(this_metrics['f1'])
    
precision_df = pd.DataFrame(np.array(precision_list))
precision_df.columns = categories

recall_df = pd.DataFrame(np.array(recall_list))
recall_df.columns = categories

f1_df = pd.DataFrame(np.array(f1_list))
f1_df.columns = categories

#crit = 1.96 
#precision_err = crit * error_margin(precision_df)
#recall_err = crit * error_margin(recall_df)
#f1_err = crit * error_margin(f1_df)


precision_err = precision_df.std(axis = 0).to_list()
recall_err = recall_df.std(axis = 0).to_list()
f1_err = f1_df.std(axis = 0).to_list()

metrics_df = pd.DataFrame({'category': categories, 'precision': precision_df.mean().to_list(), 'recall': recall_df.mean().to_list(), 'f1': f1_df.mean().to_list()})
sns.set_style("dark")
plt.rcParams["figure.figsize"] = (15,10)
ax = metrics_df.plot(kind='barh', color = ('#DDAA33', '#BB5566' , '#004488'),xerr = [precision_err, recall_err, f1_err]) #alternative color blind friendly colours ('#66c2a5', '#fc8d62' , '#8da0cb') 
ax.set_yticklabels(metrics_df.category)
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
ax.set_xlabel("Score")
ax.set_ylabel("PHROG category")

all_metrics_df = metrics_df
all_f1_err = f1_err
```
